#17.Explain the difference between apply() and map() in Pandas

Ans

Both apply() and map() are used to apply functions to data structures in Pandas, but they have distinct differences:

1.map():

Works only on Series.

Accepts a function, dictionary, or Series as input.

Applies the function element-wise.

Ideal for transforming or mapping values in a Series.

2.apply():

Works on both Series and DataFrame.

Accepts a function as input.

Applies the function along the specified axis (rows or columns in DataFrame).

Suitable for more complex operations that require aggregation or transformation across rows or columns.

#21.Why is NumPy’s array slicing faster than Python’s list slicing

Ans

NumPy's array slicing is faster due to:

1.Contiguous Memory Allocation: NumPy arrays are stored in contiguous blocks of memory, allowing for efficient access and manipulation.

2.Low-Level Implementation: NumPy operations are implemented in C, providing faster execution compared to Python's list operations.

3.No Data Copying: Slicing a NumPy array creates a view on the original data, avoiding unnecessary copying.

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
data2 = sns.load_dataset('tips')
print('data2 :',data2)
corr_matrix = data2.corr(numeric_only=True)
print('correlation matrix:',corr_matrix)
sns.heatmap(corr_matrix,annot=True)
plt.show()
```
