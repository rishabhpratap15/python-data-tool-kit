# To plot numerical values in seaborn
seaborn.relplot has options of plotting both categorical and numerical data and shows the relationship between 2 variables. also called as relationship plot

by default relplot makes scatter plot

to make line plot use parameter kind = 'line'




```python
sns.relplot(x='total_bill',y='tip',data = df)
```

```python
''' now for more analysis we use a parameter called hue , hue is used for segregation or distinguish
between some category for eg smoker, non smoker. it also gives different color coding for the
column passed.'''

sns.relplot(x='total_bill',y='tip',data = df,hue='smoker')
```

```python
# to see the size or to know how many peoples are their
sns.relplot(x='total_bill',y='tip',data = df,hue='size')
```

```python
''' for more better understanding or to draw more clear insights from the data use style inplace of hue.
it gives different representation of a column value. '''
sns.relplot(x='total_bill',y='tip',data = df,style='size')
```

```python
sns.relplot(x='total_bill',y='tip',data = df,style='smoker',hue='sex')
''' insights can be drawn from this graph for eg the most amount of tip was given by
a male who is a smoker also as the amount of bill increases the tip amount also increases.'''
```

#line plot using relplot in seaborn
by giving parameter kind = 'line'

line chart is used for data having time component and also to understand the trend in data.

```python
# to make a line chart using relplot
time = np.arange(500)
value = np.random.randint(500,size=500).cumsum()
df1 = pd.DataFrame({'time':time,'value':value})
df1
```

```python
sns.relplot(x='time',y='value',data = df1,kind='line',sort=True)
```

```python
sns.load_dataset('fmri')
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint',y='signal',data =fmri,kind='line')
```

the shaded region in the output shows that the data point at a particular point can fall between the shaded area around it.
confidence interval is its another name .


```python
sns.relplot(x='timepoint',y='signal',data =fmri,kind='line',ci=None) #  to remove confidence interval we used parameter ci = None
```

not to have confidence interval use parameter

estimator=None

```python
sns.relplot(x='timepoint',y='signal',data =fmri,kind='line',estimator=None)
```

```python
# now use of different parameters for better analysis
sns.relplot(x='timepoint',y='signal',data =fmri,kind='line',hue='event',style='region',dashes = False,marker='o')
# to remove dashes we used parameter dashes = False
```

```python
''' to understand distribution of 2 data's
1 column = univariate
2 column = bivariate
to dram bivariate plot'''
sns.jointplot(x='total_bill',y='tip',data=df,color='red')
```

jointplot understand the relationship and also distribution of 2 variables.

to see it in line chart use parameter kind = 'reg'

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg',color='purple')
```

The line depicted is called kernel density estimator (ked) it talks about the density of data for different presentation.
