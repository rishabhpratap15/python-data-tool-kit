```python
x = np.random.rand(3)
y = np.random.rand(3)
a = plt.plot(x,np.cos(y),marker='o',color='black',linewidth=4,label='cos')
b = plt.plot(y,np.sin(x),marker='o',color='yellow',linewidth=4,label='sin')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#making 2 separate plots in one page
x,y
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.cos(y),marker='o',color='pink',linewidth=4,label='cos')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,np.sin(y),marker='o',color='yellow',linewidth=4,label='sin')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# to make 4 subplots in one page
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.cos(y),marker='o',color='pink',linewidth=4,label='cos')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2,2,2)
plt.plot(x,np.sin(y),marker='o',color='yellow',linewidth=4,label='cos')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()


plt.figure()
plt.subplot(2,2,3)
plt.plot(x,np.tan(y),marker='o',color='red',linewidth=4,label='tan')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2,2,4)
plt.plot(x,np.sin(y+2),marker='o',color='green',linewidth=4,label='sin')
plt.legend()
plt.title("trigonometery graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
plt.plot(x,y,color='orange')
plt.ylim(0,8) # this changes the limit range of the axis y
plt.xlim(0,8) # this changes the limit range of the axis x
plt.axis("equal") # this is used to ensure that the plotted graph is equally distributed
plt.title("random number graph")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#stacked histogram means 2 histogram together in one page.
marks1 = np.random.rand(100)
marks2 = np.random.rand(100)
plt.hist([marks1,marks2],bins=10,color=['orange','pink'],label=['marks1','marks2'])
plt.title("marks of students")
plt.xlabel("students")
plt.ylabel("marks")
plt.legend()
plt.show()
```

```python
#pie chart is used to understand % distribution of data
subjects = ['MATHS','SCIENCE','SOCIAL SCIENCE','ENGLISH','ACCOUNTANCY']
marks = np.random.randint(100,size=(5))
plt.pie(marks,labels=subjects,autopct='%1.1f%%',shadow=True,colors=['red','orange','yellow','green','pink'],explode=(0,0,0,0.1,0)) #explode is used to sperate a value from the pie chart
plt.title("MAKRSHEET ANALYSIS OF RISHABH")
plt.xlabel("subjects")
plt.ylabel('marks')
plt.show()
```

#area plot
An area plot also called as stacked area plot is a type of data visualization
that shows how various (variables) quantitative factors have changed over time.

```python
data = [2020,2021,2022,2017,2019,2023]
year_1 = [20,23,43,44,23,13]
year_2 = [12,43,67,54,33,66]
year_3 = [12,10,9,8,56,49]
plt.stackplot(data,year_1,year_2, year_3,labels=['1st_year','2nd_year','3rd_year'],colors=['r','g','y'])
plt.legend()
plt.show()
```
