```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure,output_file,show
flowers
```

```python
output_file = ['flowers']
# to make a scatter plot
p = figure(title='flowers')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.scatter(flowers['petal_length'],flowers['petal_width'], size=3)
show(p)
```

```python
# for line chart
output_file = ['flowers']
# to make a scatter plot
p = figure(title='flowers')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.line(flowers['petal_length'],flowers['petal_width'],color='red')
show(p)
```

```python
# scatter plot using circle parameter
output_file = ['flowers']
# to make a scatter plot
p = figure(title='flowers')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'], size=3,color='purple')
show(p)
```
